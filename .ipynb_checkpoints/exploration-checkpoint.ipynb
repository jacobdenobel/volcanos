```python
import os
import operator
import functools

import obspy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import signal
%matplotlib inline
```

```python
sensors = os.listdir('data')
files = {s: sorted(list(map(lambda x: os.path.join("data", s, x), os.listdir(os.path.join("data", s))))) for s in sensors}
for sensor in sensors: 
    print(f"{sensor} n_files: {len(files[sensor])}")
    
pairs = []
for f in files['INCA']:
    for w in files['WEST']:  
        if f[-3:] == w[-3:]:
            pairs.append((f, w))
            break
print(f"We have {str(len(pairs))} pairs of data files")
```

```python
class Wrapper:
    def __init__(self, t, y):
        self.t = np.asarray(t)
        self.y = np.asarray(y)
        assert len(self.t) == len(self.y)
    
    def __add__(self, other):
        return Wrapper(np.r_[self.t, other.t], np.r_[self.y, other.y])
    
    def __iter__(self):
        return iter((self.t, self.y))

def read_file(filename:str, downsample_prc = 1) -> obspy.core.trace.Trace:
    data = obspy.read(filename)  # https://docs.obspy.org/packages/autogen/obspy.core.stream.Stream.html#obspy.core.stream.Stream
    assert len(data.traces) == 1
    trace, *_ = data.traces      # https://docs.obspy.org/packages/autogen/obspy.core.trace.Trace.html#obspy.core.trace.Trace
    assert len(trace.data) == 8640000 
    if downsample_prc:
        trace.resample(downsample_prc)            # downsample to %
    return Wrapper(trace.times("matplotlib").data, trace.data.data)

def plot_pairs(inca, west, from_=0, to=None):
    to = to or len(inca.t)
    plt.figure(figsize=(20, 10))
    assert np.all(inca.t == west.t) # check that times align
    plt.plot(inca.t[from_:to] - inca.t[0], inca.y[from_:to], label="inca")
    plt.plot(west.t[from_:to] - west.t[0], west.y[from_:to], label="west")
    plt.grid()
    plt.legend()
```

```python
data = {sensor: functools.reduce(operator.add, (read_file(f) for f in files[sensor])) for sensor in sensors}
```

```python
def plot(data, s, ax):
    d = data[s]
    ax.plot_date(*d, '-')
    ax.set_xlabel("time")
    ax.set_title("Sensor: " + s)
    ax.grid()
    
    
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(40, 20), sharex=True, sharey=True)
plot(data, 'INCA', ax1)
plot(data, 'WEST', ax2)
```

```python
def phase_shift(x1, x2, time):
    time = time - time[0]
    nsamples = len(x1)
    xcorr = signal.correlate(x1, x2)
    dt = np.r_[-time[1:][::-1], time]
    return dt[xcorr.argmax()]
```

```python
def normalize(x): return (x.max() - x) / (x.max() - x.min())
def standardize(x): return (x - np.mean(x)) / np.std(x)
def lowpass(x, freq=.5, order=8, fs=100): return signal.filtfilt(*signal.butter(order, freq, fs=fs), x)
```

```python
inca, west = pairs[0]
inca, west = read_file(inca, None), read_file(west, None)
inca.y = normalize(inca.y)
west.y = normalize(west.y)
plot_pairs(inca, west, 10000, 20000)
```

```python
inca, west = pairs[0]
inca, west = read_file(inca, None), read_file(west, None)
inca.y = standardize(inca.y)
west.y = standardize(west.y)
plot_pairs(inca, west, 10000, 20000)
```
